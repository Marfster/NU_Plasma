```python
import pandas as pd
import xlwings as xw
import numpy as np
```

```python
wb = xw.Book('Sn_L4_H8_2cycles_2.csv')
```

```python
df = wb.sheets[0].range('A1').options(pd.DataFrame, expand='table').value
```

```python
def SD2(x):
    SD2 = 2 * np.std(x)
    return SD2
```

```python
df = df.drop("Date and Time", axis=1)
df.index = df.index.astype(int)
```

```python
#df.groupby(df["Sample Name"]).agg([np.mean, SD2])
df_internal_norm_mean = df.drop("Error", axis=1)
```

```python
sub_index = df_internal_norm_mean[(df_internal_norm_mean["Sample Name"].shift(1) == df_internal_norm_mean["Sample Name"].shift(-1)) & (df_internal_norm_mean["Sample Name"].shift(1) == "NIST 200ppb")].index
#corr_values = df_internal_norm_mean[(df_external_norm_mean >= 0.74289)]
#corr_values = df_internal_norm_mean[:"20150110 10:00"]
corr_values = df_internal_norm_mean[df_internal_norm_mean["Sn118 (V)"] >= 1]
corr_values_2 = corr_values[corr_values["Sample Name"] == "NIST 200ppb"]
#corr_values = df_internal_norm_mean
sample = corr_values["Sample Name"]
sample_2 = corr_values_2["Sample Name"]
#std_mean = corr_values[(corr_values["Sample Name"] == "NIST 100ppb")].mean()
new = corr_values[corr_values.columns[corr_values.dtypes != 'object']]
new_2 = corr_values_2[corr_values_2.columns[corr_values_2.dtypes != 'object']]

#std_mean["112/120", "114/120", "115/120", "117/120", "118/120", "119/120", "122/120", "124/120"] = [0.029823, 0.020189, 0.010372, 0.235299, 0.742923, 0.263443, 0.142078, 0.177546]
#print std_mean
#epsilon =  ((new / std_mean)-1)*10000
#new["sample"] = sample
epsilon_NIST =  ((new / ((new.shift(1) + new.shift(-1))/2))-1)*10000
epsilon_2 = ((new_2 / ((new_2.shift(1) + new_2.shift(-1))/2))-1)*10000
epsilon_NIST["sample"] = sample
epsilon_2["sample"] = sample

epsilon_NIST = epsilon_NIST[new.index.isin(sub_index)]
epsilon_NIST= epsilon_NIST[epsilon_NIST["sample"] != "NIST 200ppb"]
epsilon_2 = epsilon_2[epsilon_2["sample"] == "NIST 200ppb"]
epsilon_comp = epsilon_NIST.append(epsilon_2, ignore_index=False)
#epsilon_comp.index = epsilon_comp["Date"]
#epsilon_comp.columns
epsilon_comp = epsilon_comp.sort_index(axis=0)
epsilon_comp
```

```python
def SD2(x):
    SD2 = np.nanstd(x, ddof=1) * 2
    return SD2

epsilon = epsilon_comp
#epsilon = epsilon.drop(["118"], axis = 1)
epsilon_sum = np.round(epsilon.groupby("sample").agg([np.mean, SD2]), decimals=2)
epsilon_sum
```

```python
wb.sheets[0].range('a16').value = epsilon_comp
```

```python
wb.sheets[0].range('a30').value = epsilon_sum
```
