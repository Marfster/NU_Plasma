```python
metadf = {}
```

```python
metadata_group = ["Sample", "Date", "Analysis", "Axial"]
for i in filenames:
    sample = re.search(r""+"Results_"+"(.*).txt", i)
    if sample == None:
        None
    else:
        metadf[sample.group(1)] = []
        hand = open(path_all + sample.group(0))
        for line in hand:
            for metadata_string in metadata_group:
                if metadata_string in line:
                    metadata = re.search(r""+metadata_string+".*"+": (.*)\r", line)
                    if metadata != None:
                        metadf[sample.group(1)].append(metadata.group(1))
df = pd.DataFrame.from_dict(metadf, orient = 'index')
df.columns = ["Date", "Sample", "Analysis", "Axial"]
df.sort_index(by="Date") 
```

```python
plt.plot(df[df["Sample"]=="NIST + DS 0.5-0.5 40ppb"]["Axial"], marker="o")
plt.show()
```
