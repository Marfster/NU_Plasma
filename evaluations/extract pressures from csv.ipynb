```python
metadf = {}
```

```python
metadata_group = ["Date", "Start Time", "Pir 1", "Pir 2", "Pir 3", "Pir 4", "Source Penning", "MS Pressure"]
for i in filenames:
    sample = re.search(r""+"Data_"+"(.*).csv", i)
    if sample == None:
        None
    else:
        metadf[sample.group(1)] = []
        hand = open(path_all + sample.group(0))
        for line in hand:
            for metadata_string in metadata_group:
                if metadata_string in line:
                    metadata = re.search(r""+metadata_string+",([^,]+)"+"[,\r]", line)
                    if metadata != None:
                        metadf[sample.group(1)].append(metadata.group(1))
df = pd.DataFrame.from_dict(metadf, orient = 'index')
df.columns = ["Date", "Start Time", "Pir 1", "Pir 2", "Pir 3", "Pir 4", "Source Penning", "MS Pressure"]
df.sort_index(by=["Date", "Start Time"]) 
```

```python
df.to_csv(path_all + "pressures.csv", mode='w', header=True)
```
